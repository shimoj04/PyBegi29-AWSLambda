## 1. OpenWeatherAPIから天気データを取得する
※OpenWeatherに登録しAPI_KEYを取得

- 公式サイト: [OpenWeather](https://openweathermap.org/)
- 参考URL: [世界の天気API「OpenWeatherMap」の無料APIキー発行・取得、リクエスト方法](https://auto-worker.com/blog/?p=1612)

```python
# 取得したOpenWeatherのAPI。各自別
API_KEY = '' 
```

```python
# OpenWeatherのAPIリンク
api = "http://api.openweathermap.org/data/2.5/weather?units=metric&q={city}&APPID={key}"
```

```python
# 観測する地点について
# Hokkaido, Okinawa, Tokyo, Kobeなど
city_name = ''
```

```python
# urlを作成
url = api.format(city = city_name, key = API_KEY)
print(url)
```

```python
# データを取得
response = requests.get(url)
data = response.json()
```

```python
# 取得したデータを確認
print(data)
```

```python
# 取得データの型を確認。(辞書型)
print(type(data))
```

```python
# 取得データから天気を確認
city_wea = data['weather'][0]['main']
print(city_wea)
```

```python
# 取得データから気温を確認 
city_temp = data['main']['temp']
print(city_temp)
```

```python
# 経度/緯度などを取得してみよう
```

```python
# 最低気温/最大気温を取得してみよう
```

```python
# 取得国を確認してみよう
```

```python
# 取得した地域の気温と天気を表示する文章を作ってみる。\nは改行コード
# 条件: 可変できるようにする
make_text = "{0}の気温と天気について\n気温: {1}°\n天気: {2}".format(city_name, city_temp, city_wea)
```

```python
# 表示する
print(make_text)
```

```python
# 表示されている地域名と天気を日本語にする辞書を作成
# 天気や地域について不足分あれば追加必要
area_lists = {"Okinawa": "沖縄", "Kobe": "神戸", "Hokkaido": "北海道", "Tokyo": "東京"}
weather_lists = {'Rain': '雨'}
```

```python
# リストにある場合のみ対応可能
city_name_jp = area_lists[city_name]
print(city_name_jp)
```

```python
# リストにある場合のみ対応可能
city_wea_jp = weather_lists[city_wea]
print(city_wea_jp)
```

```python
# 日本語に変換した文章を作成
make_text_jp = "{0}の気温と天気について\n気温: {1}°\n天気: {2}".format(city_name_jp, city_temp, city_wea_jp)
print(make_text_jp)
```

## 3. LineNotifyを使ってみる
※LineNofityの公式サイトより登録します。LineNotify登録できない方はwebhookを利用します

- 公式サイト: [LiINE Notify](https://notify-bot.line.me/ja/)
- 登録方法: [LINE Notifyをnode.jsで利用する](https://dev.classmethod.jp/articles/line-notify/)

- [AWS Lambdaを使ってSlackにメッセージを送る](https://qiita.com/marufeuille/items/7b513dcac1d199f8b70a)

```python
# Line Nofityのサイトに登録しアクセストークンを取得する
line_access_token = ''
```

```python
# 通知時に必要な内容を定義する
headers = {"Authorization": "Bearer %s" % line_access_token}
line_url = "https://notify-api.line.me/api/notify"
```

```python
# 1(or2)で作成した文章を辞書型に入れる
line_send_text = {'message': make_text_jp}
```

```python
# line notifyに通知する
requests.post(line_url, headers=headers, data=line_send_text)
```

### slackに通知する

```python
import slackweb
```

```python
slack = slackweb.Slack(url="聞いてください")
```

```python
# 送信
slack.notify(text=make_text_jp)
```
